This document shows some visualizations of the stock data we have obtained

```python
# import the libraries we need
import pandas as pd
import numpy as np
import datetime as dtt
import matplotlib.pyplot as plt

%matplotlib inline
```

Read in the stock data from the csv file that we have 

```python

def show(ticker, feature):
    fileName = "data/" + ticker + "/" + ticker  + "-stock-data.csv"
    dframe = pd.read_csv(fileName)
    x = pd.to_datetime(dframe['Date'])
    y = dframe[feature]
    fig = plt.figure(figsize=(18,6))
    gr = fig.add_subplot(111)
    gr.plot(x,y,'b-o')
show('AAPL', 'Short_Ratio')
show('AAPL', 'Mov_avg')
```

```python
show('GOOG', 'Short_Ratio')
show('GOOG', 'Mov_avg')
```
